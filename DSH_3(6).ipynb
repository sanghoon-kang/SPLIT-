```
import pandas as pd
import numpy as np
```

```
def make_df(cols, ind):
  data = {c: [str(c) + str(i) for i in ind]
         for c in cols}
  return pd.DataFrame(data, ind)

make_df('ABC', range(3))
```

```
#we'll create a quick class that allows us to display multiply DataFrames side by side
class display(object):
  template = """<div style="float:left; padding:10px;">
  <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
  </div>"""
  def __init__(self, *args):
    self.args = args
  def _repr_html_(self):
    return '/n'.join(self.template.format(a, eval(a)._repr_html_())
                  for a in self.args)
  def __repr__(self):
    return '/n/n'.join(a+'/n'+repr(eval(a))
                      for a in self.args)
```

```
df1 = make_df('AB', [1,2])
df2 = make_df('AB', [3,4])
pd.concat([df1, df2])
```

```
df3 = make_df('AB', [0,1])
df4 = make_df('CD', [0,1])
display('df3', 'df4', "pd.concat([df3, df4], axis=1)") #axis='col'했더니 오류 뜸
```

```
x = make_df('AB', [0,1])
y = make_df('AB', [2,3])
y.index = x.index #duplicate indices
display('x', 'y', 'pd.concat([x,y])') #pandas concatenation preserves indices, hence the 0101 repeat
```

```
try:
  pd.concat([x,y], verify_integrity=True) #if there are overlapping indices, verifity_integrity will raise an exception
except ValueError as e:
  print("ValueError:", e)
```
