```python
url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?format=csv&stationID=5415&Year={year}&Month={month}&Day={day}&timeframe=1&submit=Download+Data"
```

```python
def download_weather_month(year, month):
    url = url_template.format(day=1, year=year, month=month)
    weather_data = pd.read_csv(url, skiprows=15, index_col='Date/Time', parse_dates=True, skip_blank_lines=True)
    weather_data = weather_data.dropna(axis=1)
    weather_data.columns = [col.replace('Â', '') for col in weather_data.columns]
    weather_data = weather_data.drop(['Year', 'Day', 'Month', 'Time'], axis=1)
    return weather_data
#일년 뒤 데이터를 부르는 데에 이유가 있나? 작년 데이터가 나오는 버그? 표는 멀쩡한데
```

```python
download_weather_month(2012, 1)[:5]
```

```python
data_by_month = [download_weather_month(2012,i) for i in range(1,13)]
```

```python
weather_2012 = pd.concat(data_by_month)
weather_2012
```
