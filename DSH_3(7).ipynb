```
import pandas as pd
import numpy as np

class display(object):
  """Display HTML representation of multiple objects"""
  template = """<div style="float: left; padding: 10px;">
  <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
  </div>"""
  def __init__(self, *args):
      self.args = args
        
  def _repr_html_(self):
      return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)
    
  def __repr__(self):
      return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```
df1 = pd.DataFrame({'employee': ['Bob','Jake','Lisa','Sue'],
                   'group': ['Accounting','Engineering','Engineering','HR']})
df2 = pd.DataFrame({'employee':['Lisa','Bob','Jake','Sue'],
                   'hire_date':[2004,2008,2012,2014]})
display('df1','df2')
```

```
#one-to-one join

df3 = pd.merge(df1, df2) #automatically joins the 'employee' column
df3
```

```
#many-to-one join

df4 = pd.DataFrame({'group':['Accounting','Engineering','HR'],
                   'supervisor':['Carly','Guido','Steve']}) #preserves duplicate entries as appropriate
display('df3','df4','pd.merge(df3,df4)')
```

```
#many-to-many join

df5 = pd.DataFrame({'group':['Accounting','Accounting',
                            'Engineering','Engineering','HR','HR'],
                   'skills':['math','spreadsheets','coding','linux',
                            'spreadsheets','organization']})

display('df1','df5',"pd.merge(df1,df5)")
```

```
#'on' keyword->specify the name of the key column

display('df1', 'df2', "pd.merge(df1, df2, on='employee')")
```

```
#'left_on' and 'right_on'->when merging two datasets with different column names

df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                   'salary': [70000, 80000, 120000, 90000]})
display('df1', 'df3', 'pd.merge(df1, df3, left_on="employee", right_on="name")')
```

```
#erase redundant column

pd.merge(df1, df3, left_on="employee", right_on="name").drop('name', axis=1)
```

```
df1a = df1.set_index('employee')
df2a = df2.set_index('employee')
display('df1a', 'df2a')
```

```
#merge index

display('df1a', 'df2a',
       "pd.merge(df1a, df2a, left_index=True, right_index=True)")
```

```
# the join() method performs a merge that defaults to joining on indices

display('df1a', 'df2a', 'df1a.join(df2a)')
```

```
#mix indices and columns

display('df1a', 'df3', "pd.merge(df1a, df3, left_index=True, right_on='name')")
```

```
df6 = pd.DataFrame({'name':['Peter','Paul','Mary'],
                   'food':['fish','beans','bread']},
                  columns=['name','food'])
df7 = pd.DataFrame({'name':['Mary','Joseph'],
                   'drink':['wine','beer']},
                  columns=['name','drink'])

display('df6', 'df7', 'pd.merge(df6, df7)') #intersection (inner join)
```

```
pd.merge(df6, df7, how='inner')
```

```
display('df6', 'df7', "pd.merge(df6, df7, how='outer')")
```

```
# left join/right join->return joins over the left entries and right entries

display('df6', 'df7', "pd.merge(df6, df7, how='left')")
```

```
display('df6', 'df7', "pd.merge(df6, df7, how='right')")
```

```
# overlapping(conflicting) column names->suffixes

df8 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                    'rank':[1,2,3,4]})
df9 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'],
                   'rank':[3,1,4,2]})

display('df8','df9','pd.merge(df8,df9,on="name")') #default suffixes x, y are used
```

```
# specify custom suffixes

display('df8','df9','pd.merge(df8,df9,on="name",suffixes=["_L","_R"])')
```

```
url1 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv'
url2 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv'
url3 = 'https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv'
```

```
pop = pd.read_csv(url1)
areas = pd.read_csv(url2)
abbrevs = pd.read_csv(url3)
```

```
display('pop.head()','areas.head()','abbrevs.head()')
```

```
merged = pd.merge(pop, abbrevs, how='outer',
                 left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation',1) # This drops duplicate info
merged.head()
```

```
merged.isnull().any() #pop, state  null .
```

```
merged[merged['population'].isnull()].head()  #Puerto Rico 1990 pop 
```

```
merged.loc[merged['state'].isnull(), 'state/region'].unique() #USA state    .
```

```
#let's fill in missing info in 'state'

merged.loc[merged['state/region'] =='PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] =='USA', 'state'] = 'United States'

merged.isnull().any()
```

```
# area 

final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```
# null  .

final.isnull().any()
```

```
#area column null . 

final['state'][final['area (sq. mi)'].isnull()].unique()
```

```
# area .  null value drop.

final.dropna(inplace=True)
final.head()
```

```
#    .

data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```

```
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
```

```
density.sort_values(ascending=False, inplace=True)
density.head()
```
